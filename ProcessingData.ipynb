```python
Day_2_Nutrition_df = Preprocess_Day_2_Nutrition_df[day_2_nutrient_columns]
Day_2_Nutrition_df.head()
```

```python
Full_Days_Nutrition_df = pd.merge(Day_1_Nutrition_df, Day_2_Nutrition_df, on="SEQN")
Full_Days_Nutrition_df.head()
```

```python
for i in day_2_nutrient_columns[1:]:
    lst_i = list(i)
    lst_i[2] = '1'
    j = ''.join(lst_i)
    Full_Days_Nutrition_df[i] = Full_Days_Nutrition_df[i].fillna(value = Full_Days_Nutrition_df[j])
```

```python
Full_Days_Nutrition_df.head()
```

```python
full_day_columns = list(Full_Days_Nutrition_df.columns)
```

```python
for i in full_day_columns:
    Full_Days_Nutrition_df[i] = Full_Days_Nutrition_df[i].fillna(value = 0)
    Full_Days_Nutrition_df.loc[Full_Days_Nutrition_df[i]==0,i] = Full_Days_Nutrition_df[i].mean() 
```

```python
for i in day_1_nutrient_columns[1:]:
    day_1 = i
    lst_i = list(i)
    lst_i[2] = '2'
    day_2 = ''.join(lst_i)
    lst_i[2] = '3'
    day_3 = ''.join(lst_i)
    Full_Days_Nutrition_df[day_3] = (Full_Days_Nutrition_df[day_1] + Full_Days_Nutrition_df[day_2]) / 2
```

```python
Average_Days_Nutrition_df = Full_Days_Nutrition_df.drop(full_day_columns[1:], axis = 1)
```

```python
Average_Days_Nutrition_df.head()
```

```python
Average_Days_Nutrition_df.to_csv('2007_2008_Average_Days_Nutrition.csv')
```

```python
Preprocess_Demographics_df = pd.read_sas(filepath_or_buffer = 'C:/Users/lenovo/NHANES-Downloader-master/data/raw_data/2015-2016/Demographics/DEMO_I.XPT')
```

```python
demo_column_list = ['SEQN','RIAGENDR', 'RIDAGEYR', 'RIDRETH1','DMDHRMAR',
                   'INDFMIN2','INDFMPIR','DMDHREDU']
#demo_column_list = ['SEQN','RIAGENDR', 'RIDAGEYR', 'RIDRETH1','DMDHRMAR',
 #                       'INDFMINC','INDFMPIR','DMDHREDU']
Demographics_df = Preprocess_Demographics_df[demo_column_list]
Demographics_df.head(6)
```

RIAGENDR - Sex
RIDAGEYR - Age
RIDRETH3/RIDRETH1 - Race

```python
Demographics_df['RIAGENDR'].fillna(0, inplace = True)
Demographics_df.loc[Demographics_df['RIAGENDR']==0,'RIAGENDR'] = Demographics_df['RIAGENDR'].mode().values 
Demographics_df['RIDAGEYR'].fillna(0, inplace = True)
Demographics_df.loc[Demographics_df['RIDAGEYR']==0,'RIDAGEYR'] = Demographics_df['RIDAGEYR'].mode().values

Demographics_df['RIDRETH1'].fillna(0, inplace = True)
Demographics_df.loc[Demographics_df['RIDRETH1']==0,'RIDRETH1'] = Demographics_df['RIDRETH1'].mode().values

Demographics_df.loc[Demographics_df['DMDHRMAR'] == 77, 'DMDHRMAR'] = 5.0
Demographics_df.loc[Demographics_df['DMDHRMAR'] == 99, 'DMDHRMAR'] = 6.0
Demographics_df['DMDHRMAR'].fillna(0, inplace = True)
Demographics_df.loc[Demographics_df['DMDHRMAR']== 0,'DMDHRMAR'] = Demographics_df['DMDHRMAR'].mode().values
Demographics_df['INDFMPIR'].fillna(0,inplace = True)
Demographics_df.loc[Demographics_df['INDFMPIR']==0,'INDFMPIR'] = Demographics_df['INDFMPIR'].mean()

Demographics_df.loc[Demographics_df['INDFMIN2'] == 77, 'INDFMIN2'] = 13.0
Demographics_df.loc[Demographics_df['INDFMIN2'] == 99, 'INDFMIN2'] = 12.0
Demographics_df['INDFMIN2'].fillna(0, inplace = True)
Demographics_df.loc[Demographics_df['INDFMIN2']==0,'INDFMIN2'] = Demographics_df['INDFMIN2'].mean() 

Demographics_df['DMDHREDU'].fillna(0, inplace = True)
Demographics_df.loc[Demographics_df['DMDHREDU']== 0,'DMDHREDU'] = Demographics_df['DMDHREDU'].mode().values
```
